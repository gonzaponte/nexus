```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.size"] = 15
```

```python
att = 0.1
f   = lambda x:1 - np.exp(-x/att)

x_close = np.linspace(0.05, 0.08)
x_far   = np.linspace(0.45, 0.48)

x = np.linspace(0, 1, 101)
y = f(x)

plt.figure(figsize=(6, 5))
plt.plot(x, y, "k-")
plt.fill_between(x_close, 0, f(x_close), fc="m")
plt.fill_between(x_far, 0, f(x_far), fc="g")

plt.text(0.15, 0.2,  "low angle", c="m", transform=plt.gca().transAxes)
plt.text(0.50, 0.8, "high angle", c="g", transform=plt.gca().transAxes)

plt.xlabel("Distance travelled in material (a.u.)")
plt.ylabel("Absorption probability")
```
